How are demanding skills are trending all over the year

Methology
1- First we will convert the datetime datatype to a month format and making other column for this.
2- For percentage, will perform some actions.
3- Plot the monthly skill demand.

```python
import ast
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Now i'll convert the date to a month format and make an another column for this and group the skills with thier respective months.

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month

df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
#pivot the table for plotting a graph
df_DA_US_pivot = df_DA_US_explode.pivot_table(index="job_posted_month_no", columns='job_skills', aggfunc='size', fill_value=0)
df_DA_US_pivot
```

```python
#now we have to first sort the columns
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_US_pivot = df_DA_US_pivot.drop('Total')
df_DA_US_pivot
```

```python
df_DA_US_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
DA_totals
```

```python
df_DA_US_percent = df_DA_US_pivot.iloc[:12].div(DA_totals/100, axis=0)
df_DA_US_percent
```

```python
#change month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
from matplotlib.ticker import PercentFormatter

ax = df_DA_US_percent.iloc[:, :5].plot(kind='line', figsize=(10,6))
plt.gca().yaxis.set_major_formatter(PercentFormatter())
plt.title('Trending Top Skills for Data Analysts in the US', fontsize=14)
plt.ylabel('Percentage', fontsize=12)
plt.xlabel('Month', fontsize=12)
ax.legend().remove()
sns.despine()

for i, column in enumerate(df_DA_US_percent.iloc[:, :5].columns):
    y = df_DA_US_percent[column].iloc[-1]
    x = len(df_DA_US_percent.index) - 1
    ax.text(x, y, f' {column}', va='center',
            fontsize=9, fontweight='bold',
            color=ax.lines[i].get_color())

plt.tight_layout()
plt.show()
```
