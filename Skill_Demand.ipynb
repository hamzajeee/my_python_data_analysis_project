What are the most demanded skills for the top 3 most popular data roles?

What are we going to do:
1- First we will clean the skills row to explode them from lists.
2- After that, we will use group_by method to count the total number of skills for each role/
3- Then we will convert the counts of each skill into percentage to show on bar graph.
4- At the end, we will plot the bar graph and give insights.

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#filter data for united states
df_US = df[df['job_country'] == 'United States']
```

```python
#exploding the skills from list
df_skills = df_US.explode('job_skills')

df_skills[['job_title', 'job_skills']]
```

```python
# Counting skills per job title
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()

# Naming the new column as 'skill_count'
df_skills_count = df_skills_count.reset_index(name='skill_count')

# Sort the values by skill_count in descending order because later we will use them in plotting for top 5 skills for each respective role
df_skills_count.sort_values(by='skill_count', ascending=False, inplace=True)

df_skills_count

```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()

job_titles = sorted(job_titles[:3])

job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

sns.set_theme(style='ticks')

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short'] == job_title].head(5)[::-1]
    sns.barplot(data=df_plot, x='skill_count', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0, 45000) # make the scales the same

fig.suptitle('Counts of Skills Requested in US Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5) # fix the overlap
plt.show()
```

```python
# For representing percentages instead of skill counts on each bar, the first step would be to first calculate job_count of each job
# Use original df to get the count of job titles
df_job_title_count = df_US['job_title_short'].value_counts().reset_index(name='jobs_total')

df_job_title_count
```

```python
# Now we will merge job_title_short with df_skills_count
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left')
df_skills_perc['skill_percent'] = (df_skills_perc['skill_count'] / df_skills_perc['jobs_total']) * 100
df_skills_perc
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

sns.set_theme(style='ticks')

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)[::-1]
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='skill_count', palette='dark:b_r')
    ax[i].set_title(job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0, 78) # make the scales the same

    # remove the x-axis tick labels for better readability
    if i != len(job_titles) - 1:
        ax[i].set_xticks([])
    
    # Add percentage labels on each bar, but only if width > 0
    for p in ax[i].patches:
      width = p.get_width()
      if width > 0:
        ax[i].text(width + 1, p.get_y() + p.get_height() / 2, f'{width:.1f}%', va='center')

    

fig.suptitle('Percentages of Skills Requested in US Job Postings', fontsize=15)
fig.tight_layout(h_pad=0.5) # fix the overlap
plt.show()

```
